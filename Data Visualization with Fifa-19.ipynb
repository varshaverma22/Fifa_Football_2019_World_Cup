```python
# Bar Chart are another useful visualization tool for analyzing categories in data.
# We find the Most common nationalities found in our fifa-19 
# Count
# Collections group the data
```

```python
from collections import Counter  # It is count the number of times Nationality
print(Counter(df["Nationality"]))# constructure
```

```python
# We can filter this dictionary using the most_common method
# Let's look are the 10 most common nationality value
print(dict(Counter(df['Nationality']).most_common(10)))
```

```python
# To analyze most infrequent nationality values.
least_common= Counter(df['Nationality']).most_common()[::-1]
print(least_common[0:10])
```

```python
print(least_common[0:30]) # it provide the least common naltionality
```

```python
# Finally, to genrate the bar plot of the 10 most common nationality values,
# We simple call 
```

```python
nationality_dict= dict(Counter(df['Nationality']).most_common(10)) # 10 most common nationality
plt.bar(nationality_dict.keys(),nationality_dict.values())
plt.xlabel("Nationality") # label the x axis
plt.ylabel("Frequency") # label the y axis
plt.title("Bar plot of ten Most Common Nationality") # It provide the Title
plt.show()
```

```python
# We Rotate the Nationality
nationality_dict= dict(Counter(df['Nationality']).most_common(10)) # count the 10 most commmon nationality 
plt.bar(nationality_dict.keys(),nationality_dict.values())
plt.xlabel("Nationality")
plt.ylabel("Frequency")
plt.title("Bar plot of ten Most Common Nationality")
plt.xticks(rotation=90) # xtrick method is used to rotate the Nationality
plt.show()
```

```python
# Least Common
nationality_dict= dict(least_common[0:30]) # take the least common nationality
plt.bar(nationality_dict.keys(),nationality_dict.values())
plt.xlabel("Nationality")
plt.ylabel("Frequency")
plt.title("Bar plot of ten Most Common Nationality")
plt.xticks(rotation=90) # Rotation of the graph
plt.show()
```

```python
# Pie Chart are a useful way to visualize proportions in your data
# Visualize the proportions of player 
# Nationality_2 # This is the new Column
# ~ this indicate the not
# key name of country
df.loc[df.Nationality=='England','Nationality_2']="England"
df.loc[df.Nationality=='Germany','Nationality_2']="Germany"
df.loc[df.Nationality=='Spain','Nationality_2']="Spain"
df.loc[df.Nationality=='Argentina','Nationality_2']="Argentina"
df.loc[~df.Nationality.isin(['England','Germany','Spain','Argentina']),'Nationality_2']="Other"
# Create the Dictionary That contain the proportion value foreach of these:
proportion=dict(Counter(df['Nationality_2']))
for key, values in proportion.items():
    proportion[key]=(values)/len(df)*100   # It convert into percentage value
print(proportion)
```

```python
# Highlight the Country through the explode.
# We have a five value so we take the 5 value
explode=(0,0,0,0,0.3) # 0,0 is centre distance form the centre
# It show the values and keys
plt.pie(proportion.values(),labels=proportion.keys(),autopct='%1.2f%%', explode=explode)
plt.figure(figsize=(10,7))
plt.show() 
```

```python
# To regenrate our histrogram of the overall column
# We use the distplot method on the seaborn object:
import seaborn as sns
sns.distplot(df['Overall']) # Distplot provide the distribution plot and so as
# the naturethe plot wheather distribution is normal, right skewd and left suide
plt.xlabel('Overall')
plt.ylabel('Frequency')
plt.title("Histrogram of Overall Rating")
plt.grid()
plt.show()
```

```python
# Seaborn also maks generating scatter plot plots straightforward
# Let's recreate the scatter plot from earlier.
sns.scatterplot(df['Overall'],df['Wage'])
plt.title('Overall vs. Wage')
plt.xlabel('Overall')
plt.ylabel('Wage')
plt.grid()
plt.show()
```

```python
# First you make the Heat map after it you make the scatter plot see the correlation
# Heat Map  (Important)
# It clear dependencies
# Seaborn is also for making correlation heatmap, which can be used to identify variable
# Correlation work on integer
# To genrate the correlation between a set of numerical columns
# age, Overall, Wage and Skill moves:
corr=df[['Overall','Age','Wage','Skill Moves']].corr()   # It find the corrrelation
sns.heatmap(corr,annot=True) # we set annot to true to see the correlation value
plt.title('Heatmap of Overall, Age, wage_euro, Skill Moves')
plt.show()
# Dark color show low vaue
# Take low diagonal and upper diagonal
# 1 indicate high correaltion 
```

```python
# Pair Plot (Important)
# This allow you to genrate a matrix of distribution and scatter plots for a set of numerical features.
# It is a quick and easy way to visualize both the distribution in numerical values and 
# relationship between variables through scatterplot
data=df[['Overall','Age','Potential']]
sns.pairplot(data) # It plot the pairplot of the data
plt.show()
```
